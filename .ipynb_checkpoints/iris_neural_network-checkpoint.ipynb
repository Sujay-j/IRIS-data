```python
import pandas as pd
from pandas import get_dummies
import numpy as np
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.python.keras import utils
import matplotlib.pyplot as plt
from keras.utils import to_categorical
import seaborn as sn
from sklearn.model_selection import train_test_split
```

```python
names = ['sepalLength','sepalWidth','petalLength','petalWidth','Species']
data = pd.read_csv('iris.data',names = names , header = 0)
data.head(5)
```

```python
data['Species']=data['Species'].astype("category")
data.dtypes
```

```python
data.isnull().sum()
```

```python
corrmatt = data[['sepalLength','sepalWidth','petalLength','petalWidth','Species']].corr()
mask = np.array(corrmatt)
mask[np.tril_indices_from(mask)] = False
fig,ax = plt.subplots()
fig.set_size_inches(20,10)
sn.heatmap(corrmatt,mask=mask,vmax =.8,square = True,annot = True)
```

```python
from sklearn.preprocessing import LabelBinarizer
species_lb = LabelBinarizer()
Y = species_lb.fit_transform(data.Species.values)
```

```python
output_data = data["Species"]
input_data = data.drop("Species",axis=1)
x_train,x_test,y_train,y_test=train_test_split(input_data,Y,test_size = 0.3,random_state = 42)
x_train.shape[0]
```

class irisclassifier(Model):
    def __init__(self):
        super(irisclassifier,self).__init__()
        self.layer1 = Dense(10,input_dim = 4,activation ='relu')
        self.layer2 = Dense(10,activation='relu')
        self.outputlayer = Dense(3,activation='softmax')
        
    def call(self, x):
        x = self.layer1(x)
        y = self.layer2(y)
        return self.outputlayer(x)
    
model = irisclassifier()

model.compile(optimizer = tf.keras.optimizers.Adam(),loss = 'categorical_crossentropy',metrics = ['accuracy'])

```python
model=Sequential()
model.add(Dense(10,input_dim = 4,activation ='relu'))
model.add(Dense(10,activation ='relu'))
model.add(Dense(3,activation ='softmax'))
model.compile(optimizer = tf.keras.optimizers.Adam(),loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=300,batch_size=10)
```

```python
scores = model.evaluate(x_test,y_test)
print("\nAccuracy : %.2f%%" %(scores[1]*100))
```

```python
prediction = model.predict(x_test)
prediction1 = pd.DataFrame({'IRIS1':prediction[:,0],'IRIS2':prediction[:,1], 'IRIS3':prediction[:,2]})
prediction1.round(decimals=4).head()
```
